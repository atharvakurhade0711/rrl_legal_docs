```python
import json
```

```python
f = open('test_data.json')

data = json.load(f)
```

```python
data[0].keys()
```

```python
for i in range(len(data)):
    org_text = data[i]['data']['text']
    new_text = ""
    rrl_list = data[i]['annotations'][0]['result']
    last_end=0

    for rrl_token in rrl_list:
        text = rrl_token['value']['text']

        
        start = org_text.lower().find(text.lower())
        
        end  = start + len(text)
        label = rrl_token['value']['labels'][0]
        new_text += '<' + label + '>' + ':{ ' + org_text[start:end] + ' } '
        last_end = end
    
    data[i]['rrl_embedded_text'] = new_text
```

```python
new_data = []
```

```python
for i in range(len(data)):
    org_text = data[i]['data']['text']
    new_text = ""
    ner_list = data[i]['ner_annotations'][0]['result']
    last_end=0
    dic =  {}
    dic['id'] = data[i]['id']
    for ner_token in ner_list:
        text = ner_token['value']['text']
        start = org_text.lower().find(text.lower())
        end  = start + len(text)
        label = ner_token['value']['labels'][0]
        new_text +=  org_text[last_end:start] + '<' + label + '>' + org_text[start:end] + '</' + label + '>'
        last_end = end
    
    #new_text += '<NONE>' + org_text[last_end:len(org_text)-1] + '</NONE>'
    dic['data'] = {}
    dic['data']['text'] = new_text
    new_data.append(dic)
```
